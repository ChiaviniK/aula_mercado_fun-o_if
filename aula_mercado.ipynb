```python
# o nosso desafio é mostrar os produtos de um mercado
# quero poder escolher os produtos, para ver infos
# quero poder adicionar novos produtos
# quero poder excluir um produto do mercado

#----------

# criar as variaveis globias para o software
lista_produtos = ['Detergente ype',
                  'Refrigerante Gengibirra',
                  'Arroz',
                  'Feijão',
                  'Água mineral']

lista_preco = [2.50, 5, 7.25, 4, 3.45]
```

```python
# para criarmos uma função
# para criarmos um apanhado de código, devemos
# armazenar eles dentro de um bloco e dar um nome

# função 1 - ver produtos
def ver_produtos():
  print(f'Os produtos que temos são: {lista_produtos}')
  print(f'e seus preços são: {lista_preco}')

def adicionar_produtos():
  novo_produto = input('Informe o nome do produto: ')
  novo_preco = float(input('Informe o preço: '))
  lista_produtos.append(novo_produto)
  # /\ CHAMAMOS A LISTA - ATRIBUIMOS A FUNÇÃO APPEND (ADICIONA) -
  # COLOCAMOS O QUE VAI SER ADICIONADO
  lista_preco.append(novo_preco)
```

```python
print('-----MENU DO MERCADO-----')
print('Escolha a opção:')
print('1 - ver produtos')
print('2 - cadastrar novo produto')
opcao = int(input('informe a opção desejada'))

# o comando if serve para condicionar algo
# if condição operador(==) resposta:
if opcao == 1:
  # o que ele deve executar
  ver_produtos()
# o comando elif é a combinação do termos else (se nao) if (se)
elif opcao == 2:
  adicionar_produtos()
  ver_produtos()
else:
  print('opçõa nao encontrada')
```
